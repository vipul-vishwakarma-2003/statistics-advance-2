```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

s_mean = 150
ppl_mean = 145
std = 10
s_size = 30
alpha = 0.5

z_stat = (s_mean - ppl_mean) / (std / math.sqrt(s_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

s_mean = 150
ppl_mean = 145
std = 10
s_size = 30
alpha = 0.5

z_stat = (s_mean - ppl_mean) / (std / math.sqrt(s_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

critical_value = stats.norm.ppf(1 - alpha / 2)

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, loc=0, scale=1)

plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color="red", alpha=0.5, label="Rejection Region")

plt.axvline(x=-critical_value, color='green', linestyle='--', label=f"Critical Z = {-critical_value:.2f}")
plt.axvline(x=critical_value, color='green', linestyle='--', label=f"Critical Z = {critical_value:.2f}")
plt.axvline(x=z_stat, color='orange', linestyle='-', label=f"Calculated Z = {z_stat:.2f}")

plt.title("Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def visualize_errors(mu_null, mu_alt, sigma, n, alpha):
    se = sigma / np.sqrt(n)
    z_alpha = stats.norm.ppf(1 - alpha)
    crit_value = mu_null + z_alpha * se

    p_type_1_error = 1 - stats.norm.cdf(crit_value, mu_null, se)

    z_beta = (crit_value - mu_alt) / se
    p_type_2_error = stats.norm.cdf(z_beta)

    power = 1 - p_type_2_error

    print(f"Type 1 Error (α): {p_type_1_error:.4f}")
    print(f"Type 2 Error (β): {p_type_2_error:.4f}")
    print(f"Power of the test: {power:.4f}")

    x = np.linspace(mu_null - 4*sigma, mu_alt + 4*sigma, 1000)
    null_dist = stats.norm.pdf(x, mu_null, se)
    alt_dist = stats.norm.pdf(x, mu_alt, se)

    plt.figure(figsize=(10, 6))

    plt.plot(x, null_dist, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H1)', color='red')

    plt.fill_between(x, null_dist, where=(x >= crit_value), color='gray', alpha=0.5, label="Type 1 Error (α) Region")
    plt.fill_between(x, alt_dist, where=(x < crit_value), color='orange', alpha=0.5, label="Type 2 Error (β) Region")

    plt.axvline(x=crit_value, color='black', linestyle='--', label="Critical Value")

    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.xlabel('Test Statistic')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

mu_null = 0
mu_alt = 2
sigma = 1
n = 30
alpha = 0.05

visualize_errors(mu_null, mu_alt, sigma, n, alpha)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

before = np.array([72, 75, 68, 70, 74, 71, 69, 73, 70, 72])
after = np.array([70, 72, 65, 68, 71, 70, 67, 72, 68, 70])

t_statistic, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the paired groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the paired groups.")

plt.figure(figsize=(8, 6))
sns.pointplot(x=np.repeat(['Before', 'After'], len(before)),
              y=np.concatenate([before, after]),
              hue=np.tile(range(len(before)), 2),
              palette='viridis', linestyles='--', markers='o')
plt.title("Paired Plot: Before vs After")
plt.ylabel("Measurement")
plt.xlabel("Condition")
plt.show()

mean_before = np.mean(before)
mean_after = np.mean(after)
std_error_before = stats.sem(before)
std_error_after = stats.sem(after)

plt.figure(figsize=(6, 6))
plt.bar(['Before', 'After'], [mean_before, mean_after],
        yerr=[std_error_before, std_error_after],
        capsize=10, color=['skyblue', 'lightgreen'])
plt.title("Comparison of Means: Before vs After")
plt.ylabel("Mean Measurement")
plt.show()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import scipy.stats as st
import numpy as np

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = st.sem(data)
    margin_of_error = std_err * st.t.ppf((1 + confidence) / 2, n - 1)

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [12, 15, 14, 10, 13, 17, 16, 11, 18, 14]
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

def margin_of_error(sample_std_dev, sample_size, confidence_level):
    alpha = 1 - confidence_level
    z_score = stats.norm.ppf(1 - alpha / 2)

    standard_error = sample_std_dev / math.sqrt(sample_size)

    margin_error = z_score * standard_error
    return margin_error

sample_std_dev = 12.5
sample_size = 100
confidence_level = 0.95

moe = margin_of_error(sample_std_dev, sample_size, confidence_level)
print(f"Margin of Error: {moe:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python

from scipy.stats import chi2_contingency

data = np.array([[50, 30],
                 [20, 40]])
table = pd.DataFrame(data, columns=["Like", "Dislike"], index=["Male", "Female"])

chi2, p, dof, expected = chi2_contingency(table)

print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(pd.DataFrame(expected, columns=["Like", "Dislike"], index=["Male", "Female"]))

alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: Variables are dependent (association exists).")
else:
    print("Fail to reject the null hypothesis: No significant association.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

def calculate_expected_frequencies(observed):
    observed = np.array(observed)
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total
    return expected

observed_data = [
    [50, 30, 20],
    [30, 50, 20],
    [20, 30, 50]
]

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Expected Frequencies:")
print(expected_frequencies)
```

```python
#15. Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics

def plot_chi_square_distribution(df_values, num_samples=10000, bins=50):
    x = np.linspace(0, max(df_values) * 3, 1000)

    plt.figure(figsize=(10, 6))

    for df in df_values:
        data = np.random.chisquare(df, num_samples)
        plt.hist(data, bins=bins, density=True, alpha=0.5, label=f'χ² df={df}')

        plt.plot(x, stats.chi2.pdf(x, df), lw=2)

    plt.title("Chi-square Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

df_values = [1, 2, 5, 10, 20]
plot_chi_square_distribution(df_values)
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples

np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=5, size=30)
sample2 = np.random.normal(loc=10, scale=7, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(sample1) - 1
df2 = len(sample2) - 1

p_value = 2 * (1 - stats.f.cdf(F_stat, df1, df2))

print(f"F-statistic: {F_stat:.4f}")
print(f"p-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.diagnostic import het_breuschpagan

def check_anova_assumptions(data, dv, iv):
    model = ols(f'{dv} ~ C({iv})', data=data).fit()
    residuals = model.resid

    print("### Checking Assumptions for ANOVA ###\n")

    shapiro_test = stats.shapiro(residuals)
    print(f"Shapiro-Wilk Test for Normality:\n W={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")
    if shapiro_test.pvalue > 0.05:
        print("Residuals are normally distributed.\n")
    else:
        print("Residuals may not be normally distributed.\n")

    dw_stat = durbin_watson(residuals)
    print(f"Durbin-Watson Test for Independence: {dw_stat:.4f}")
    if 1.5 < dw_stat < 2.5:
        print("No strong evidence of autocorrelation.\n")
    else:
        print("Possible autocorrelation detected.\n")

    groups = [group[dv].values for name, group in data.groupby(iv)]
    levene_test = stats.levene(*groups)
    print(f"Levene’s Test for Equal Variance:\n W={levene_test.statistic:.4f}, p-value={levene_test.pvalue:.4f}")
    if levene_test.pvalue > 0.05:
        print("Homogeneity of variance is met.\n")
    else:
        print("Possible heteroscedasticity detected.\n")

    plt.figure(figsize=(8, 6))
    plt.scatter(model.fittedvalues, residuals, alpha=0.5)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel("Fitted Values")
    plt.ylabel("Residuals")
    plt.title("Residuals vs. Fitted Values (Check for Independence & Homoscedasticity)")
    plt.show()

if __name__ == "__main__":
    np.random.seed(42)
    df = pd.DataFrame({
        "Group": np.repeat(["A", "B", "C"], 10),
        "Score": np.concatenate([np.random.normal(50, 10, 10),
                                 np.random.normal(55, 10, 10),
                                 np.random.normal(60, 10, 10)])
    })

    check_anova_assumptions(df, "Score", "Group")
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

from statsmodels.graphics.factorplots import interaction_plot

np.random.seed(42)

factor_A = np.repeat(['Low', 'High'], 30)
factor_B = np.tile(['Type1', 'Type2', 'Type3'], 20)

data_values = (
    5 + 2 * (factor_A == 'High') + 3 * (factor_B == 'Type2') +
    4 * (factor_B == 'Type3') + np.random.normal(0, 1, 60)
)

df = pd.DataFrame({'Factor_A': factor_A, 'Factor_B': factor_B, 'Value': data_values})

model = ols('Value ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-Way ANOVA Results:")
print(anova_table)

plt.figure(figsize=(8, 6))
interaction_plot(df['Factor_A'], df['Factor_B'], df['Value'], colors=['red', 'blue', 'green'])
plt.title('Interaction Plot of Factor_A and Factor_B')
plt.xlabel('Factor_A')
plt.ylabel('Mean Value')
plt.show()
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

np.random.seed(42)

sample_size = 100
mean1, std1 = 50, 10
mean2, std2 = 52, 10

sample1 = np.random.normal(mean1, std1, sample_size)
sample2 = np.random.normal(mean2, std2, sample_size)

t_stat, p_value = stats.ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

sample_size = 20
sample_variance = 5.5
hypothesized_variance = 4
alpha = 0.05

chi2_statistic = (sample_size - 1) * sample_variance / hypothesized_variance
df = sample_size - 1
p_value = 1 - stats.chi2.cdf(chi2_statistic, df)
critical_value = stats.chi2.ppf(1 - alpha, df)

print(f"Chi-square Test Statistic: {chi2_statistic:.4f}")
print(f"Critical Value at α={alpha}: {critical_value:.4f}")
print(f"P-value: {p_value:.4f}")

if chi2_statistic > critical_value:
    print("Reject H0: There is evidence that the population variance is greater than 4 grams².")
else:
    print("Fail to reject H0: There is not enough evidence to say the variance is greater than 4 grams².")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, size_a, success_b, size_b):
    p1 = success_a / size_a
    p2 = success_b / size_b
    p_pool = (success_a + success_b) / (size_a + size_b)
    se = np.sqrt(p_pool * (1 - p_pool) * (1/size_a + 1/size_b))
    z_score = (p1 - p2) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_value

success_a = 50
size_a = 200
success_b = 30
size_b = 180

z_score, p_value = z_test_proportions(success_a, size_a, success_b, size_b)
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the proportions.")
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

F = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(data1) - 1
df2 = len(data2) - 1

p_value = 1 - stats.f.cdf(F, df1, df2)

alpha = 0.05

print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

plt.figure(figsize=(12, 5))

sns.kdeplot(data1, label='Dataset 1', fill=True, alpha=0.5)
sns.kdeplot(data2, label='Dataset 2', fill=True, alpha=0.5)
plt.axvline(np.mean(data1), color='blue', linestyle='dashed', label="Mean Dataset 1")
plt.axvline(np.mean(data2), color='red', linestyle='dashed', label="Mean Dataset 2")
plt.title('Comparison of Two Distributions')
plt.legend()
plt.show()
```
